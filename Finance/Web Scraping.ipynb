```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average"
```

```python
const = pd.read_html(url)[1]
```

```python
const = const.iloc[:, :5].copy()
```

```python
const
```

```python
const.rename(columns = {"Date added": "date_added"}, inplace = True)
```

```python
const.head()
```

```python
const.info()
```

```python
const.date_added = pd.to_datetime(const.date_added)
```

```python
const.info()
```

```python
tickers = const.Symbol.to_list()
```

## Loading and Saving Historical Stock Prices

```python
import yfinance as yf
```

```python
df = yf.download(tickers, start = "2007-01-01", end = "2020-03-31")
```

```python
df = df.loc[:, "Close"].copy()
```
