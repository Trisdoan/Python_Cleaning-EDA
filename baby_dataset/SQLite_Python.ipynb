# Importing & merging many Files (complex case)

6. __Load__ all files (*.txt) and __merge/concatenate__ all files into one Pandas DataFrame.

```python
pd.read_csv("CA.TXT", header = None, names = ["states","gender", "year","name" , "count"])
```

```python
dfs = []
for file in file_names:
    df = pd.read_csv("{}".format(file), header = None, names = ["states","gender", "year","name" , "count"])
    dfs.append(df)
```

```python
dfs[0]
```

7. Create a __RangeIndex__ and __save__ the DataFrame (with columns "State", "Gender", "Year", "Name", "Count") in a new csv-file.

```python
data = pd.concat(objs = dfs, axis =0).reset_index()
```

```python
data.drop(columns = "index",axis = 1)
```

```python
data.to_csv("baby_names_states.csv", index=False)
```
