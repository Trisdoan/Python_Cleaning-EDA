## Statistical Analysis and Hypothesis Testing with scipy

In the follwing work with __Ranks__!
Check whether __GDP__ (Standard of Living), __Total_Games__ (Political Stability measure) and __Population__ (Size) have an effect on __Total Medals__. (hint: work with spearman correlation, not with pearson correlation)

```python
medals_per_country.drop(columns = ["Gold", "Silver","Bronze"], inplace = True)
```

```python
medals_per_country.corr(method = "spearman")
```

```python
import scipy.stats as stats
```

## Hypothesis 1: no relationship between Total medals and Population

```python
r, p_value = stats.spearmanr(medals_per_country.Total, medals_per_country.Population, nan_policy = "omit")
```

```python
print("Critical value is: " + str(r))

print("P value is: " + str(p_value))
```

## Hypothesis 2: no relationship between Total medals and GDP per capita

```python
r, p_value = stats.spearmanr(medals_per_country.Total, medals_per_country["GDP per Capita"], nan_policy = "omit")
```

```python
print("Critical value is: " + str(r))

print("P value is: " + str(p_value))
```

## Hypothesis 3: no relationship between Total medals and Participations

```python
r, p_value = stats.spearmanr(medals_per_country.head(50).Total, medals_per_country["Total_Games"].head(50), nan_policy = "omit")
```

```python
print("Critical value is: " + str(r))

print("P value is: " + str(p_value))
```

## Aggregating and Ranking

Create the following __Seaborn Heatmap__ with __Medal Ranks__ for Top 50 Countries (Total Medals, Summer Games Medals, Winter Games Medals, Men, Women).

```python
sum_vs_win = pd.crosstab(index = olympics.Country, columns = olympics.Edition)
sum_vs_win
```

```python
men_vs_women = pd.crosstab(index = olympics.Country, columns = olympics.Gender)
men_vs_women
```

```python
medal_by_cat = pd.concat([men_vs_women, sum_vs_win], axis =1 )
medal_by_cat["Total"] = medal_by_cat.Summer.add(medal_by_cat.Winter)
medal_by_cat.sort_values(by = "Total", ascending = False, inplace = True)
```

```python
ranks = medal_by_cat.rank(ascending = False, method = "average")
ranks
```

```python
top_50 = ranks.head(50)
top_50
```

```python
plt.figure(figsize = (50,10))
sns.heatmap(top_50.T)
plt.show()
```
